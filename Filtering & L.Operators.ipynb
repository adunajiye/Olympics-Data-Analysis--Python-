```python
# Import Data
athlete = pd.read_csv('C:/Users/HP/Videos/Machine Learning/Data science/Pandas/Projects/athlete_events.csv')
region = pd.read_csv('C:/Users/HP/Videos/Machine Learning/Data science/Pandas/Projects/noc_regions.csv')

athlete.head(5)
```

```python
# Join the dataframes on Noc 
Athelete_df = pd.merge(athlete,region, how = 'left', on = 'NOC')
Athelete_df.head(5)
```

```python
# Make Columns name Consistent by replacing columns 
Athelete_df.rename(columns =
 {'region':'Region','notes':'Notes'
 },inplace = True)
Athelete_df.head(2)
```

```python
# Check null values 
Athelete_df.info()
```

```python
# Checking the total numbers of null values in the dataframe 
Athelete_df.isna().value_counts().unique()
check_df = Athelete_df.isnull().sum()
check_df.sort_values(ascending = False)
```

```python
# India,Japan & Nigeria Details 
India_details = Athelete_df.query('Team == "India"')
India_details.head(4)

# 2. Japan 
Japan_details = Athelete_df.query('Team == "Japan"')
Japan_details.head(2)

# 3. Nigeria 
Nigeria_details =Athelete_df.query('Team == "Nigeria"')
Nigeria_details.head(10)

```

```python
# Top 10 countries participating & its Percentage 
Top_10_Countries = Athelete_df['Team'].value_counts().sort_values(ascending = False).head(10)
Top_10_Countries.head(10)
```

```python
# Percentage of Top 10 countries 
Pct_Top_10_Countries =  Athelete_df['Team'].value_counts(normalize = True)
Pct_Values  = Pct_Top_10_Countries * 100
Pct_Values.head(10)
```

```python
# Concat Values 
Concated_Top_10  = pd.concat([Top_10_Countries,Pct_Values],axis = 'columns',sort = False)
Concated_Top_10.head()
```

```python
# Rename Concated Columns 
Concated_Top_10.rename (columns={'Team':'Amount_Of_Occurence',
'Team':'Percentage (%) Occurence'
})
```
